```python
from gensim.models import Word2Vec, KeyedVectors

ko_model = Word2Vec.load('/gdrive/MyDrive/colab/ko/ko.bin')
```

```python
kor_v = ko_model.wv["한국"]
seoul_v = ko_model.wv["서울"]
paris_v = ko_model.wv["파리"]
print(ko_model.wv.similar_by_vector(kor_v - seoul_v + paris_v))
```

```python
com_v = ko_model.wv["컴퓨터"]
sc_v = ko_model.wv["과학"]
humanities_v = ko_model.wv["인문학"]
print(ko_model.wv.similar_by_vector(com_v - sc_v + humanities_v))
```

```python
love_v = ko_model.wv["사랑"]
lost_v = ko_model.wv["이별"]
print(ko_model.wv.similar_by_vector(love_v + lost_v))
```

```python
cn_v = ko_model.wv["충청북도"]
cj_v = ko_model.wv["청주"]
gw_v = ko_model.wv["강원도"]

print(ko_model.wv.similar_by_vector(cn_v - cj_v + gw_v))
```

```python
human_v = ko_model.wv["인간"]
print(ko_model.wv.similar_by_vector(human_v - love_v))
```

```python
art_v = ko_model.wv["미술"]
paint_v = ko_model.wv["물감"]
music_v = ko_model.wv["음악"]

print(ko_model.wv.similar_by_vector(art_v - paint_v + music_v))
```
